```python
import numpy as np
from matplotlib import pyplot as plt
import os
from tqdm import tqdm
```

```python
data_path = '/work/gg0302/g260141/NACHMO_data/dyn_OH'
```

```python
def load_data(data_path):
    files = os.listdir(data_path)
    data = dict()
    for file in tqdm(files):
        if file.endswith('.npy'):
            data[file[:-4]] = np.load(os.path.join(data_path, file))
    return data
```

```python
data = load_data(data_path)
```

```python
t = data['Time_oh'][0] * 60.0

species = [d for d in data.keys() if not d.startswith('Time')]
species
```

```python
data['OH'].shape
```

```python
def plot_sim(data, i, logy=True):    
    ax = []
    for ind_s, s in enumerate(species):
        ax.append(plt.subplot(len(species), 1, ind_s + 1))
        if logy:
            plt.semilogy(t, data[s][i], label=s)
        else:
            plt.plot(t, data[s][i], label=s)
        plt.ylabel(f'{s} conc.')
        if s == len(species) - 1:
            plt.xlabel('Time [min]')
    return ax
    #plt.legend()
```

```python
ax = plot_sim(data, np.random.randint(10000))
for a in ax:
    a.set_xlim([-.05, 1])
```
